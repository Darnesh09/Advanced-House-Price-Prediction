```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',None)
```

```python
data = pd.read_csv("/home/darnesh/Documents/programming/Machine Learning/house-prices-advanced-regression-techniques/train.csv")
data.head()
```

```python
cat_features_nan = [feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes==object]
for feature in cat_features_nan:
    print(f"{feature} has {np.round(data[feature].isnull().mean()*100,2)}% of nan values")
```

```python
for feature in cat_features_nan:
    data[feature] = data[feature].fillna('Missing')
```

```python
data[cat_features_nan].sample(5)
```

```python
num_feature_nan = [feature for feature in data.columns if data[feature].isnull().sum()>0 and data[feature].dtypes!=object]
for feature in num_feature_nan:
    print(f"{feature} has {np.round(data[feature].isnull().mean()*100,2)}% of nan values")
```

```python
for feature in num_feature_nan:
    data[feature+'_nan'] = np.where(data[feature].isna(),1,0)
    data[feature].fillna(data[feature].median(),inplace=True)

data[num_feature_nan]
```

```python
data.sample(6)
```

another way to impute missing values using sklearn's SimpleImputer is given below

```python
df = data.copy()
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
df[num_feature_nan] = imputer.fit_transform(df[num_feature_nan])
df[num_feature_nan]
```

```python
imputer.statistics_
```

when we have many outliers we should replace them with median or mode but not with mean

```python
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    data[feature] = data['YrSold']-data[feature]

data[['YearBuilt','YearRemodAdd','GarageYrBlt']].sample(6)
```

```python
temp = data.groupby('KitchenQual')['SalePrice'].count()/len(data)
temp
```

```python
temp_df = temp[temp>0.03].index
temp_df
```

```python
# numerical features with all non zero values
for feature in ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']:
    data[feature] = np.log(data[feature])

data[['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].sample(6)
```

## Handling rare categorical features
remove categorical variables that are present less than 1% in the observation

```python
cat_features = [feature for feature in data.columns if data[feature].dtypes=='object']
cat_features
```

```python
for feature in cat_features:
    temp = data.groupby(feature)['SalePrice'].count()/len(data)
    temp = temp[temp>0.01].index
    data[feature] = np.where(data[feature].isin(temp),data[feature],'Rare_val')
```

```python
data[cat_features].sample(6)
```

```python
for feature in cat_features:
    order = data.groupby(feature)['SalePrice'].mean().sort_values().index
    order_value = {k:i for i,k in enumerate(order)}
    data[feature] = data[feature].map(order_value)

data.sample(5)
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

feature_scale = [feature for feature in data.columns if feature not in ['Id','SalePrice'] and feature not in cat_features]

scaled = scaler.fit_transform(data[feature_scale])

data[feature_scale] = scaled

data.head()
```
